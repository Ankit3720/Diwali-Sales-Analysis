```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
#ro avoid encoding error, use 'unicode_escape'
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
#drop unreleated/blamk columns
df.drop(['Status', 'unnamed1'], axis= 1, inplace= True)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace= True)
#drop null values
```

```python
pd.isnull(df).sum()
```

```python
df['Amount'] = df['Amount'].astype('int')
#change data type
```

```python
df['Amount'].dtype
#check data type
```

```python
df.columns
```

```python
df.rename(columns= {'Marital_Status':'Shaddi'})
#rename column
```

```python
df.describe()
```

```python
df[['Age', 'Orders', 'Amount']].describe()
#use describe() for specific columns
```

```python
df.columns
```

```python
ax= sns.countplot(x= 'Gender', data= df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen= df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False)

sns.barplot(x= 'Gender', y= 'Amount', data= sales_gen)
```

```python
ax= sns.countplot(x= 'Age Group',hue= 'Gender', data= df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen= df.groupby(['Age Group'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False)

sns.barplot(x= 'Age Group', y= 'Amount', data= sales_gen)
```

```python
sales_gen= df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by= 'Orders', ascending= False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x= 'State', y= 'Orders', data= sales_gen)
```

```python
sales_gen= df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x= 'State', y= 'Amount', data= sales_gen)
```

```python
ax= sns.countplot(x= 'Marital_Status', data= df)

sns.set(rc={'figure.figsize':(3,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state= df.groupby(['Marital_Status', 'Gender'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x= 'Marital_Status', y= 'Amount', hue= 'Gender', data= sales_state)
```

```python
ax= sns.countplot(x= 'Occupation', data= df)

sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen= df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x= 'Occupation', y= 'Amount', data= sales_gen)
```

```python
ax= sns.countplot(x= 'Product_Category', data= df)

sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen= df.groupby(['Product_Category'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x= 'Product_Category', y= 'Amount', data= sales_gen)
```

```python
sales_gen= df.groupby(['Product_ID'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x= 'Product_ID', y= 'Amount', data= sales_gen)
```
